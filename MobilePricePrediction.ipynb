```python
import pandas as pd
import numpy as np

test = pd.read_csv('test.csv')
print(test.shape)
test.head()
```

```python
train = pd.read_csv('train.csv')
print(train.shape)
train.head()
```

```python
print(train.dtypes)
```

```python
all_features = train.drop(['price_range'], axis = 1).values
all_classes = train['price_range'].values
```

```python
from sklearn import svm
from sklearn.model_selection import cross_val_score

model = svm.SVC(kernel = 'linear', C = 1.0).fit(all_features, all_classes)

all_features_test = test.drop(['id'], axis = 1).values
predictions = model.predict(all_features_test)
predictions
```
