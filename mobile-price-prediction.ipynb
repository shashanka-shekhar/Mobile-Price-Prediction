```python
import pandas as pd
import numpy as np

train = pd.read_csv(r'../input/mobile-price-classification/train.csv')
train.head()
```

```python
train.info()
```

```python
test = pd.read_csv(r'../input/mobile-price-classification/test.csv')
test.info()
```

```python
features = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width',
           'ram', 'sc_h', 'sc_w', 'talk_time']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

for i in features:
    k = np.array(train[i])
    k = k.reshape(-1, 1)
    K = scaler.fit_transform(k)
    train[i] = K
```

```python
train.head()
```

```python
X = train.drop('price_range', axis = 1)
y = train['price_range']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier(n_estimators = 1000)

model.fit(X_train, y_train)

predictions = model.predict(X_test)

ac = accuracy_score(predictions, y_test)

ac
```

```python
for i in features:
    k = np.array(test[i])
    k = k.reshape(-1, 1)
    K = scaler.fit_transform(k)
    test[i] = K
```

```python
test_copy = test
```

```python
Test = test.drop('id', axis = 1)
```

```python
Test.info()
```

```python
test_output = model.predict(Test)
```
